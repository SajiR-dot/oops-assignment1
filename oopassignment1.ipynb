```python
# Question 1

# Bank Account Create a class representing a bank account with attributes like account number,account holder name and balance implement methods to deposit and withdraw money from the account 

import logging as log
log.basicConfig(filename="bank_account_log.log",level=log.DEBUG)
log.info("Creating a bank Account class and initializing it with account_number account_name , account_balance attributes")
class Bank_account:
    def __init__(self,account_number,account_name):
        self.account_number = account_number
        self.account_name = account_name
        self.account_balance = 10000


    log.info("defining method deposit_money")
    def deposit_money(self,amount):
        try:
            print(f"Your account balance is {self.account_balance+amount} Rupees")
            print(f"The amount {amount} Rupees was deposited successfully")
        except Exception as e:
            log.error(f"The program hit with an exception {e} ")
            print(e)


    log.info("Creating a method to withdraw_money")
    def withdraw_money(self,amount):
        log.info("The money was withdrawn")
        print(f"You have withdrew {amount} Rupees and your account balance is {self.account_balance-amount} Rupees ")

log.info("Creating an object from the class Bank Account")
account1 = Bank_account(1,"sajeeth")
print(account1.account_name)

account1.deposit_money("100")

account1.withdraw_money(1000)
```

```python
# Question 3

# Vehicle rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle , return a vehicle and display available vehicles

import logging as log
log.basicConfig(filename="vehicle_rental.log",level=log.DEBUG)

log.info("Creating a class VehicleRental")
class VehicleRental:
    def __init__(self):
        self.vehicles = {"ford":1,"porsche":2,"ferrari":3,"bmw":5,"kawasaki":4,"royal enfield":3}
    log.info("Creating a method rent which accepts the vehicle type as parameter")
    def rent(self,vehicle_type):
        try:
            log.info("The type of vehicle is available")
            print(f"{self.vehicles[vehicle_type]} available")
            if self.vehicles[vehicle_type] >= 1:
                log.info("The vehicle can be rented")
                print(f"You can rent the {vehicle_type}")
                self.vehicles[vehicle_type] -= 1
            else:
                log.info("The vehicle is not available")
                self.vehicles.pop(vehicle_type)
                print("Sorry you can rent it is unavailable")
        except Exception as e:
            log.error("There was an Key Error")
            print(f"{e} type vehicles are not available here")

    log.info("Creating a method to return the vehicle")
    def veh_return(self,vehicle_type):
        log.info("The vehicle was returned successfully")
        print(f"Your {vehicle_type} was returned successfully")
        self.vehicles[vehicle_type]+=1

    log.info("Creating a method to display the available vehicles")
    def available_veh(self):
        log.info("The available vehicles are printed")
        print("The available vehicles are")
        for key,value in self.vehicles.items():
            print(key,value)

rental_1 = VehicleRental()
rental_1.available_veh()
rental_1.rent("porsche")
rental_1.available_veh()
rental_1.veh_return("porsche")
rental_1.available_veh()
```

```python
# Question 4

# library catalog Create a class representing a library and a book. Implement methods to add books to the library, borrow books and display available books

import logging as log
log.basicConfig(filename="library.log",level=log.DEBUG)

log.info("Creating a class book with attributes name, genre and id")
class Book:
    def __init__(self,book_name,book_genre,book_id):
        self.name = book_name
        self.genre = book_genre
        self.id = book_id

log.info("Creating a class library and inheriting the book class")
class Library(Book):
    books = []

    log.info("Creating a method to add book which accepts the book name as the parameter")
    def add_book(self,book_name):
        log.info("The book was added to the list")
        self.books.append(book_name)
        print("Your book was added to the library")

    log.info("Creating a method to borrow a book")
    def borrow(self,book_name):
        log.info("The book was successfully borrowed")
        self.books.remove(book_name)
        print(f"you have borrowed {book_name} ")

    log.info("Creating a method to display books")
    def display_books(self):
        if len(self.books)>=1:
            log.info("The books were displayed")
            print("The available books are")
            print(self.genre)
            print(self.id)
            print(self.books)
        else:
            log.info("The books are not available")
            print("Currently there are no books available")

lib1 = Library("12 rules for life","self help",1)

lib1.add_book("12 rules for life")
lib1.add_book("karma")
print(lib1.books)
lib1.borrow("12 rules for life")
lib1.display_books()
```

```python
# Question 5

# Product inventory create classes representing a product and an inventory system Implement methods to add products to the inventory , update product quantity and display available products

import logging as log
log.basicConfig(filename="Inventory.log",level=log.DEBUG)

log.info("creating a product class")
class Product:
    def __init__(self,product_name,product_quantity):
        self.name = product_name
        self.quantity = product_quantity

log.info("Creating a inventory class")
class Inventory:
    products = {}  # create a empty dictionary to store the products information
    def add_products(self,product_name,product_quantity): # def a method to add the products
        self.products[product_name] = product_quantity
        log.info("Product was added successfully")
        print("Your product was added successfully")

    def display(self):
        for key,values in self.products.items():
            print(f" {key} = {values}kg")

    def update_quantity(self,name,quantity):
        self.products[name]+=quantity
        print("The product was updated")

product1 = Product("salt",1) # The products can be created from the Product class
product2 = Product("rice",10)
inventory = Inventory() # The inventory can be created from the Inventory class

inventory.add_products(product1.name,product1.quantity) # Add products is used to add the products to the inventory
inventory.add_products(product2.name,product2.quantity)
inventory.display()
inventory.update_quantity("rice",2) #   The update quantity is used to update the products
inventory.display()
inventory.update_quantity("salt",2)
inventory.display()
```

```python
# Question 6

# Shape calculation Create a class representing a shape with attributes like length, width and height. Implement methods to calculate the area and the perimeter of the shape


import logging as log
log.basicConfig(filename="shape.log",level=log.DEBUG)

log.info("Creating a class shape")
class Shape:
    def __init__(self,length,width,height=0):
        self.shape_length = length
        self.shape_width = width
        self.shape_height = height
    log.info("Creating a method to calculate the area")
    def calc_area(self,shape_name:str):
        if shape_name.lower()=='square':
            print(f"The area of square is {self.shape_length**2}")
            log.info("The area of square was calculated")
        elif shape_name.lower()=='rectangle':
            print(f"The area of {shape_name} is {self.shape_length * self.shape_width}")
            log.info("The area of rectangle was calculated")
    log.info("A method is defined to calculate the perimeter")
    def perimeter(self, shape_name):
        if shape_name.lower() == 'square':
            print(f"The perimeter of square is {self.shape_length * 4}")
            log.info("The perimeter of square was calculated")
        elif shape_name.lower() == 'rectangle':
            print(f"The perimeter of {shape_name} is {2 * (self.shape_length + self.shape_width)}")
            log.info("The perimeter of rectangle was calculated")


square = Shape(10,10,10)
square.calc_area("Square")
shape2 = Shape(20,32)
shape2.calc_area("rectangle")
shape2.perimeter("square")
```

```python
# Question 7

# Student management Create a class representing a student with attributes like student ID ,name and grades.Implement method to calculate the average grade and display student details


import logging as log

log.basicConfig(filename="student.log", level=log.DEBUG)

log.info("Creating a class student")
class Student: # define a student class
    def __init__(self,id,name, **grades): # using the init constructor initialize the id, name , using kwargs accept the grades
        self.student_id = id
        self.student_name = name
        self.student_grades = grades
    log.info("Craating a method to calculate average grade")
    def average_grade(self): # This method is used to calculate average grade

        print(f"The average is {round(sum(self.student_grades.values())/len(self.student_grades))}")
        log.info("The average grade was calculated")

    log.info("Creating a method called student details")
    def student_details(self): # This method is used to display the students

        print(f"The student id is '{self.student_id}'")
        print(f"The student name is '{self.student_name}'")
        print(f"The student grades are '{self.student_grades}'")
        log.info("Student details were displayed ")

student1 = Student(1,"saji",maths = 100, science=70,hindi=60)
student1.average_grade()
student1.student_details()
student2 = Student(2,"Gokul",maths=100,science=80,hindi=70,sst=90)
student2.average_grade()
student2.student_details()
```

```python
# Question 9

# Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implements methods to add posts, display posts and search for posts by keyword


import logging as log
log.basicConfig(filename="profile_log.log",level=log.DEBUG)

log.info("The program starts here im creating a class called profile")
class Profile:
    log.info("creating attribute user_name with the init constructor and defining an attribute posts as empty dictionary")
    def __init__(self, user_name):
        self.user_name = user_name
        self.posts = {}
    log.info("Creating a method to add post")
    def add_post(self,keyword,post):
        self.posts[keyword] = post
        print("Your post was added successfully")
        log.info("Post was added successfully")

    log.info("Creating a method to display posts")
    def display_posts(self):
        for key,value in self.posts.items():
            print(f"Tag = {key}, Post = {value}")
        print(f"This was posted by {self.user_name}")  
        log.info("Post was displayed successfully")

    log.info("creating a method to search the posts")
    def search_post(self,keyword):
        try:
            print(self.posts[keyword])
        except KeyError:
            log.error("There was a key Error")
            print("There was a key error")


log.info("An object was created from the class profile")
profile_1 = Profile("SAJI")
profile_1.add_post("life","This life is very precious and awesome to me")
profile_1.display_posts()
profile_2 = Profile("Gokul")
profile_2.add_post("Bangalore","life on mars")
profile_2.display_posts()
profile_1.search_post(1234)
```

```python
# Question 10 

# Todo List Create a class representing a Todo list with attributes like tasks and due dates. Implement methods to add tasks , mark tasks as completed, and display pending tasks.

import logging as log

log.basicConfig(filename="todo_list_log",level=log.DEBUG)

log.info("Creating the class todo")
class Todo:
    tasks_counter = []
    def __init__(self,task,due_date):
        self.task = task
        self.due_date = due_date

    log.info("The method add_task is created")
    def add_task(self):
        self.tasks_counter.append(self.task)
        print("Your Task has been added")

    log.info("The method mark_completed is created")
    def mark_completed(self):
        self.tasks_counter.clear()

    log.info("The method for displaying the pending the remaining tasks is created")
    def display_pending(self):
        if self.tasks_counter==[]:
            print("There are no pending tasks")
        else:
            print(self.tasks_counter,self.due_date)

log.info("The object from the todo class is created")
todo_task1 = Todo("drink 1 litre of water","12/10/23")
todo_task1.add_task()
todo_task2 = Todo("Eat 10 almonds","22/10/23")
todo_task2.add_task()


```
